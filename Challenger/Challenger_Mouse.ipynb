```python
# Load packages
import requests 
import pandas as pd 
import time

from bs4 import BeautifulSoup 
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.common.by import By
```

```python
current_url = 'https://www.challenger.sg/search?query=mouse&page=2' 
```

```python
resp = requests.get(current_url)
content = BeautifulSoup(resp.content, 'lxml')
print(content)
```

```python
def scrape_Challenger(url):
    service = Service(executable_path="/chromedriver")
    driver = webdriver.Chrome(service=service)
    driver.get(url)

    content = driver.find_elements(By.CLASS_NAME,'search-item-box')
    mouse_list = []

    for post in content:
         try:
            data = {
            "title": post.find_element(By.CLASS_NAME,'product-name').text,
            "final_price": post.find_element(By.CLASS_NAME,'product-price').text,
            # "discount_price": post.find_element(By.CLASS_NAME,'special-price').text,
            # "price": post.find_element(By.CLASS_NAME,'old-price').text,
            # "discount": post.find_element(By.CLASS_NAME,'price-save-amount').text,
            # "Desc": post.find_element(By.CLASS_NAME,'navision-spec').text
            }
         except:
            continue

         mouse_list.append(data)
    return pd.DataFrame(mouse_list)
```

```python
df_mouse = scrape_Challenger(current_url)
df_mouse.to_csv("Challenger_Mouse.csv", index= False)
df_mouse
```
