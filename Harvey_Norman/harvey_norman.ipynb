```python
# Load packages
import requests 
import pandas as pd 
import time

from bs4 import BeautifulSoup 
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.common.by import By
```

```python
url_list = []

for i in range(1,3):
    url = f"https://www.harveynorman.com.sg/index.php?subcats=Y&status=A&pshort=N&pfull=N&pname=Y&pkeywords=Y&search_performed=Y&q=laptop&dispatch=products.search&page={i}"
    url_list.append(url)

print(url_list)
```

```python
item_list = []
def scrape_product_details(url):
    service = Service(executable_path="/chromedriver")
    driver = webdriver.Chrome(service=service)

    driver.get(url)
    content = driver.find_elements(By.CSS_SELECTOR,'.hproduct-col.product-col.clearfix')    
 
    for post in content:
        data = {
                  "title": None,
                  "Price": None,
                  "Availability": None,
            }
        try:
            data["title"] = post.find_element(By.CSS_SELECTOR,'div.bx > div > a').text
        except:
            pass
        try:
            data["Price"] = post.find_element(By.CLASS_NAME, 'price').text
        except:
            pass
        try:
            data["Availability"] = post.find_element(By.CLASS_NAME, 'price-old').text
        except:
            pass
        try:

        item_list.append(data)
    return pd.DataFrame(item_list)
```

```python
for url in url_list:
    df = scrape_product_details(url)
    
df.to_csv("harvey_norman.csv", index=False)
df
```
