```python
# Load packages
import requests 
import pandas as pd 
import time

from bs4 import BeautifulSoup 
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.common.by import By
```

```python
genres = ["travel_2","mystery_3","historical-fiction_4"]
url_list = []

for genre in genres:
    url = f"http://books.toscrape.com/catalogue/category/books/{genre}/index.html"
    url_list.append(url)


print(url_list)
```

```python
books_list = []
def scrape_books_details(url):
    service = Service(executable_path="/chromedriver")
    driver = webdriver.Chrome(service=service)

    driver.get(url)
    content = driver.find_elements(By.CSS_SELECTOR,'.col-xs-6.col-sm-4.col-md-3.col-lg-3')    
 
    for post in content:
        data = {
                  "title": None,
                  "Price": None,
                  "Availability": None,
            }
        try:
            data["title"] = post.find_element(By.CSS_SELECTOR,'article > h3 > a').text
        except:
            pass
        try:
            data["Price"] = post.find_element(By.CSS_SELECTOR, 'article > div.product_price > p.price_color').text
        except:
            pass
        try:
            data["Availability"] = post.find_element(By.CSS_SELECTOR, 'article > div.product_price > p.instock.availability').text
        except:
            pass


        books_list.append(data)
    return pd.DataFrame(books_list)
```

```python
for url in url_list:
    df = scrape_books_details(url)
    
# df_jobs.to_csv("courts_keyboards.csv", index=False)
df
```
