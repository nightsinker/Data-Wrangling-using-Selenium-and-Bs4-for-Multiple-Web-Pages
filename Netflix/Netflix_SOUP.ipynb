```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site = 'https://en.wikipedia.org/wiki/Netflix'
response = requests.get(base_site)
response
```

```python
html = response.content
html[:100]
```

```python
soup = BeautifulSoup(html,"html.parser")
```

```python
soup
```

```python
link = soup.find_all('head')
link
```
