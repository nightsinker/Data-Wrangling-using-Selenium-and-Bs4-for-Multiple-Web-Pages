```python
resp = requests.get(current_url)
content = BeautifulSoup(resp.content, 'lxml')
print(content)
```

```python
def scrape_books_details(url):
    service = Service(executable_path="/chromedriver")
    driver = webdriver.Chrome(service=service)

    driver.get(url)
    content = driver.find_elements(By.CLASS_NAME,'search-result__item ')    
    books_list = []
    for post in content:
        data = {
                  "title": None,
                  "author": None,
                  "Cover": None,
                  "Publish_Date":None,
                  "Price": None,
            }
        try:
            data["title"] = post.find_element(By.CSS_SELECTOR,'div.product-info > h4 > a').text
        except:
            pass
        try:
            data["author"] = post.find_element(By.CLASS_NAME, 'product-author').text
        except:
            pass
        try:
            data["Cover"] = post.find_element(By.CSS_SELECTOR, 'div.product-info > p.product-format > span:nth-child(1)').text
        except:
            pass
        try:
            data["Publish_Date"] = post.find_element(By.CSS_SELECTOR, 'div.product-info > p.product-format > span:nth-child(3)').text
        except:
            pass
        try:
            data["Price"] = post.find_element(By.CSS_SELECTOR, 'div.product-info > div > ul > li.product-price--current').text
        except:
            pass

        books_list.append(data)
    return pd.DataFrame(books_list)
```

```python
df = scrape_books_details(current_url)
# df_jobs.to_csv("blackwell.csv", index=False)  
df
```
