```python
Keyboard_list = []
def scrape_Courts(url):
    service = Service(executable_path="/chromedriver")
    driver = webdriver.Chrome(service=service)
    driver.get(url)

    content = driver.find_elements(By.CLASS_NAME,'product-item-info')
    
    for post in content:

      data = {
                  "title": None,
                  "final_price": None,
                  "old_price": None,
                  "discount": None,
                  "Desc": None
            }
      try:
            data["title"] = post.find_element(By.CLASS_NAME, 'product-item-link').text
      except:
            pass

      try:
            data["final_price"] = post.find_element(By.CLASS_NAME, 'price').text
      except:
            pass

      try:
            data["old_price"] = post.find_element(By.CLASS_NAME, 'old-price').text
      except:
            pass

      try:
            data["discount"] = post.find_element(By.CLASS_NAME, 'price-save-amount').text
      except:
            pass

      try:
            data["Desc"] = post.find_element(By.CLASS_NAME, 'navision-spec').text
      except:
            pass

      Keyboard_list.append(data)
    return pd.DataFrame(Keyboard_list)
```

```python
for url in current_url:
    df_jobs = scrape_Courts(url)

# df_jobs = scrape_Courts(current_url)
    
df_jobs.to_csv("courts_keyboards.csv", index=False)
df_jobs
```
